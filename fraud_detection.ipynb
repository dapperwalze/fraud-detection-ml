```python
from sklearn.preprocessing import StandardScaler

#Normalize the amount column
scaler = StandardScaler()
df['Amount_scaled'] = scaler.fit_transform(df[['Amount']])

#Drop Time and original Amount - not needed for training
df_clean = df.drop(columns = ['Time', 'Amount'])

print("Original df shape: ", df.shape)
print("Cleaned df shape: ", df_clean.shape)
print("\nFirst 3 rows of cleaned data:")
df_clean.head(3)
```

```python
# X = everything the model learns FROM
# Y = what the model is trying to predict
X = df_clean.drop(columns= ['Class'])
Y = df_clean['Class']

print("X shape:", X.shape)
print("Y shape:", Y.shape)
print("\nX columns:", X.columns.tolist())
print("\ny value counts")
print(Y.value_counts())
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 42, stratify = Y)

print("X_train shape:", X_train.shape)
print("X_test shape:", X_test.shape) 
print("\nFraud in training set:", Y_train.sum())
print("\nFraud in test set:", Y_test.sum())
print("\nFraud rate in train:", round(Y_train.mean() * 100, 3), "%")
print ("Fraud rate in test:", round(Y_test.mean() * 100, 3), "%")
```

```python
print("=== Day 3 Summary ===")
print(f"Features (X): {X_train.shape[1]} columns")
print(f"Training rows: {X_train.shape[0]:,}")
print(f"Test rows: {X_test.shape[0]:,}")
print(f"Fraud in train: {Y_train.sum()} ({Y_train.mean()*100:.3f}%)")
print(f"Fraud in test: {Y_test.sum()} ({Y_test.mean()*100:.3f}%)")
print(f"\nData is ready for modeling.")
```

```python
from sklearn.linear_model import LogisticRegression
import time

print("Training Logisitic Regression...")
start = time.time()

model_lr = LogisticRegression(max_iter=1000, random_state=42)
model_lr.fit(X_train, Y_train)

end = time.time()
print(f"Training complete in {round(end - start, 2)} seconds")
print(f"\nModel trained on {X_train.shape[0]:,} transactions")
print(f"Using {X_train.shape[1]} features")
```

```python
# Make predictions on the test set

y_pred = model_lr.predict(X_test)
y_prob = model_lr.predict_proba(X_test)[:, 1]

print("Sample predictions (first 10):")
print("Predicted:", y_pred[:10])
print("Actual:   ", Y_test.values[:10])

print(f"\nTotal predicted frauds: {y_pred.sum()}")
print(f"Total actual frauds:    {Y_test.sum()}")
```

```python
from sklearn.metrics import accuracy_score

accuracy = accuracy_score(Y_test, y_pred)
print(f"Accuracy: {round(accuracy * 100, 3)}%")
print(f"\nThink about this:")
print(f"A model that predicts All transactions as legitimate would be:")
dummy_accuracy = (Y_test == 0).sum() / len(Y_test)
print(f"{round(dummy_accuracy * 100, 3)}% accurate")
print(f"\nOur model improvement over dummy: {round((accuracy - dummy_accuracy) * 100, 3)}%")
```

```python
print("=== Day 4 Summary ===")
print(f"Model: Logistic Regression")
print(f"Training time: 1.74 seconds")
print(f"Accuracy: {round(accuracy_score(Y_test, y_pred) * 100, 3)}%")
print(f"Predicted frauds: {y_pred.sum()}")
print(f"Actual frauds: {Y_test.sum()}")
print(f"Missed frauds: {Y_test.sum() - y_pred.sum()}")
print(f"\nAccuracy is misleading. Full evaluation happens on Day 5.")
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

cm = confusion_matrix(Y_test, y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=['Legitimate', 'Fraud'])
disp.plot(cmap='Blues')
plt.title('Confusion Matrix - Logistic Regression')
plt.show()

print("Raw confusion matrix:")
print(cm)
```

```python
from sklearn.metrics import classification_report, roc_auc_score

print("=== Classification Report ===")
print(classification_report(Y_test, y_pred, target_names=['Legitimate', 'Fraud']))

roc_auc = roc_auc_score(Y_test, y_prob)
print(f"ROC AUC Score: {round(roc_auc, 4)}")
```

```python
from sklearn.metrics import roc_curve

fpr, tpr, thresholds = roc_curve(Y_test, y_prob)

plt.figure(figsize=(7, 5))
plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'ROC Curve (AUC = 0.956)')
plt.plot([0, 1], [0, 1], color='navy', lw=1, linestyle='--', label='Random guess')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate (Recall)')
plt.title('ROC Curve — Logistic Regression')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
print("=== Day 5 Summary — Logistic Regression Scorecard ===")
print(f"Confusion Matrix:")
print(f"  True Negatives  (Legit correctly ignored): 56,851")
print(f"  False Positives (Legit wrongly flagged):       13")
print(f"  False Negatives (Fraud missed):                35")
print(f"  True Positives  (Fraud correctly caught):      63")
print(f"\nPrecision:  83%   — when we flag fraud, we're right 83% of the time")
print(f"Recall:     64%   — we caught 64% of all actual fraud")
print(f"F1-Score:   0.72  — balance of precision and recall")
print(f"ROC AUC:    0.956 — excellent ranking ability")
print(f"\nTarget for Day 6: improve Recall above 80%")
```

```python
from imblearn.over_sampling import SMOTE

print("Applying SMOTE to balance training data...")
sm = SMOTE(random_state=42)
X_res, Y_res = sm.fit_resample(X_train, Y_train)

print(f"Before SMOTE - Fraud: {Y_train.sum()}/ Total: {len(Y_train):,}")
print(f"After SMOTE - Fraud: {Y_res.sum()}/ Total: {len(Y_res):,}")

print("\nRetraining Gradient Boosting on balanced data...")
start = time.time()

model_gb = GradientBoostingClassifier(n_estimators=100, random_state=42)
model_gb.fit(X_res, Y_res)

end = time.time()
print(f"Training complete in {round(end - start, 2)} seconds")

y_pred_gb = model_gb.predict(X_test)
y_prod_gb = model_gb.predict_proba(X_test)[:,1]

print("\n=== Gradient Boosting + SMOTE Results ===")
print(classification_report(Y_test, y_pred_gb, target_names=['Legitimate', 'Fraud']))
print(f"ROC AUC: {round(roc_auc_score(Y_test, y_prob_gb), 4)}")
```
