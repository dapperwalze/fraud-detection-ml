```python
from sklearn.preprocessing import StandardScaler

#Normalize the amount column
scaler = StandardScaler()
df['Amount_scaled'] = scaler.fit_transform(df[['Amount']])

#Drop Time and original Amount - not needed for training
df_clean = df.drop(columns = ['Time', 'Amount'])

print("Original df shape: ", df.shape)
print("Cleaned df shape: ", df_clean.shape)
print("\nFirst 3 rows of cleaned data:")
df_clean.head(3)
```

```python
# X = everything the model learns FROM
# Y = what the model is trying to predict
X = df_clean.drop(columns= ['Class'])
Y = df_clean['Class']

print("X shape:", X.shape)
print("Y shape:", Y.shape)
print("\nX columns:", X.columns.tolist())
print("\ny value counts")
print(Y.value_counts())
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 42, stratify = Y)

print("X_train shape:", X_train.shape)
print("X_test shape:", X_test.shape) 
print("\nFraud in training set:", Y_train.sum())
print("\nFraud in test set:", Y_test.sum())
print("\nFraud rate in train:", round(Y_train.mean() * 100, 3), "%")
print ("Fraud rate in test:", round(Y_test.mean() * 100, 3), "%")
```

```python
print("=== Day 3 Summary ===")
print(f"Features (X): {X_train.shape[1]} columns")
print(f"Training rows: {X_train.shape[0]:,}")
print(f"Test rows: {X_test.shape[0]:,}")
print(f"Fraud in train: {Y_train.sum()} ({Y_train.mean()*100:.3f}%)")
print(f"Fraud in test: {Y_test.sum()} ({Y_test.mean()*100:.3f}%)")
print(f"\nData is ready for modeling.")
```

```python
from sklearn.linear_model import LogisticRegression
import time

print("Training Logisitic Regression...")
start = time.time()

model_lr = LogisticRegression(max_iter=1000, random_state=42)
model_lr.fit(X_train, Y_train)

end = time.time()
print(f"Training complete in {round(end - start, 2)} seconds")
print(f"\nModel trained on {X_train.shape[0]:,} transactions")
print(f"Using {X_train.shape[1]} features")
```

```python
# Make predictions on the test set

y_pred = model_lr.predict(X_test)
y_prob = model_lr.predict_proba(X_test)[:, 1]

print("Sample predictions (first 10):")
print("Predicted:", y_pred[:10])
print("Actual:   ", Y_test.values[:10])

print(f"\nTotal predicted frauds: {y_pred.sum()}")
print(f"Total actual frauds:    {Y_test.sum()}")
```

```python
from sklearn.metrics import accuracy_score

accuracy = accuracy_score(Y_test, y_pred)
print(f"Accuracy: {round(accuracy * 100, 3)}%")
print(f"\nThink about this:")
print(f"A model that predicts All transactions as legitimate would be:")
dummy_accuracy = (Y_test == 0).sum() / len(Y_test)
print(f"{round(dummy_accuracy * 100, 3)}% accurate")
print(f"\nOur model improvement over dummy: {round((accuracy - dummy_accuracy) * 100, 3)}%")
```

```python
print("=== Day 4 Summary ===")
print(f"Model: Logistic Regression")
print(f"Training time: 1.74 seconds")
print(f"Accuracy: {round(accuracy_score(Y_test, y_pred) * 100, 3)}%")
print(f"Predicted frauds: {y_pred.sum()}")
print(f"Actual frauds: {Y_test.sum()}")
print(f"Missed frauds: {Y_test.sum() - y_pred.sum()}")
print(f"\nAccuracy is misleading. Full evaluation happens on Day 5.")
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

cm = confusion_matrix(Y_test, y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=['Legitimate', 'Fraud'])
disp.plot(cmap='Blues')
plt.title('Confusion Matrix - Logistic Regression')
plt.show()

print("Raw confusion matrix:")
print(cm)
```

```python
from sklearn.metrics import classification_report, roc_auc_score

print("=== Classification Report ===")
print(classification_report(Y_test, y_pred, target_names=['Legitimate', 'Fraud']))

roc_auc = roc_auc_score(Y_test, y_prob)
print(f"ROC AUC Score: {round(roc_auc, 4)}")
```

```python
import xgboost as xgb
import time
from sklearn.metrics import classification_report, roc_auc_score

#Calculate the imbalance ratio
neg = (Y_train == 0).sum()
pos = (Y_train == 1).sum()
scale = neg / pos

print(f"Legitimate: {neg:,}")
print(f"Fraudulent: {pos}")
print(f"scale_pos_weight: {round(scale, 1)}")

print("\nTraining XGBoost...")
start = time.time()

model_xgb = xgb.XGBClassifier(n_estimators = 100, scale_pos_weight = scale, random_state=42, eval_metric='logloss', verbosity=0)
model_xgb.fit(X_train, Y_train)

end = time.time()
print(f"Training complete in {round(end - start, 2)} seconds")

#Evaluate 
Y_pred_xgb = model_xgb.predict(X_test)
Y_prob_xgb = model_xgb.predict_proba(X_test)[:,1]

print("\n=== XGBoost Results ===")
print(classification_report(Y_test, Y_pred_xgb, target_names=['Legitimate', 'Fraud']))
print(f"ROC AUC: {round(roc_auc_score(Y_test, Y_prob_xgb), 4)}")
```

```python
import mlflow
import mlflow.sklearn
import mlflow.xgboost
from sklearn.metrics import precision_score, recall_score, f1_score, roc_auc_score

#Set the experiment name - all runs will live under this
mlflow.set_experiment("fraud-detection")

print("MLflow experiment set: fraud-detection")
print("Tracking URI:", mlflow.get_tracking_uri())
```

```python
# Define all four models and their results
experiments = [
    {
        "name": "Logistic Regression",
        "model": model_lr,
        "params": {"model_type": "logistic_regression", "max_iter": 1000},
        "preds": y_pred,
        "probs": y_prob,
        "log_fn": mlflow.sklearn.log_model
    },
    {
        "name": "Random Forest",
        "model": model_rf,
        "params": {"model_type": "random_forest", "n_estimators": 100},
        "preds": y_pred_rf,
        "probs": y_prob_rf,
        "log_fn": mlflow.sklearn.log_model
    },
    {
        "name": "RF + SMOTE",
        "model": model_rf_smote,
        "params": {"model_type": "random_forest_smote", "n_estimators": 100, "smote": True},
        "preds": y_pred_rf_smote,
        "probs": y_prob_rf_smote,
        "log_fn": mlflow.sklearn.log_model
    },
    {
        "name": "XGBoost",
        "model": model_xgb,
        "params": {"model_type": "xgboost", "n_estimators": 100, "scale_pos_weight": 577.3},
        "preds": Y_pred_xgb,
        "probs": Y_prob_xgb,
        "log_fn": mlflow.xgboost.log_model
    }
]

for exp in experiments:
    with mlflow.start_run(run_name=exp["name"]):
        # Log parameters
        mlflow.log_params(exp["params"])

        # Log metrics
        mlflow.log_metrics({
            "precision": round(precision_score(Y_test, exp["preds"]), 4),
            "recall": round(recall_score(Y_test, exp["preds"]), 4),
            "f1": round(f1_score(Y_test, exp["preds"]), 4),
            "roc_auc": round(roc_auc_score(Y_test, exp["probs"]), 4)
        })

        # Log model artifact
        exp["log_fn"](exp["model"], name="model")

        print(f"Logged: {exp['name']}")

print("\nAll runs logged successfully")
```
